```python
import pandas as pd
import numpy as np

np.random.seed(42)  # For reproducibility

num_students = 100
num_questions = 10
mean_marks = 6.5
std_dev = 2.0

# Generate normal distributed marks and clip to 0-10 range
marks = np.random.normal(loc=mean_marks, scale=std_dev, size=(num_students, num_questions))
marks = np.clip(marks, 0, 10)  # Ensure marks within 0 and 10

# Round marks
marks = np.round(marks, 1)

# Calculate total marks
total_marks = np.sum(marks, axis=1)

# Create dataframe with student_id, question marks and total_marks
student_ids = [f"S{i:03d}" for i in range(1, num_students + 1)]
df = pd.DataFrame(marks, columns=[f"Q{q}" for q in range(1, num_questions + 1)])
df.insert(0, "student_id", student_ids)
df["total_marks"] = total_marks

# Show sample data
print(df.head(10))

# Save to CSV if needed
df.to_csv("student_marks_normal_dist.csv", index=False)
```

```python
import numpy as np

np.random.seed(42)  # For reproducibility

num_samples = 1000
mean = 50
std_dev = 15

# Generate normal distributed values
continuous_vals = np.random.normal(loc=mean, scale=std_dev, size=num_samples)

# Round to get discrete values
discrete_vals = np.round(continuous_vals).astype(int)

# Clip values to range 0 to 100
discrete_vals = np.clip(discrete_vals, 0, 100)

print(discrete_vals[:20])  # Preview first 20 discrete normal values
```
